```python
df_country=pd.read_excel('Country-Code.xlsx')
df_country.head()
```

```python
df_country.columns
```

```python
final_df=pd.merge(df,df_country, on='Country Code', how='left')
```

```python
final_df.head()
```

```python
final_df.dtypes
```

```python
final_df.columns
```

```python
#Taking country names in final_df
country_name=final_df.Country.value_counts().index
```

```python
#Taking values for no of register outlets
country_values=final_df.Country.value_counts().values
```

```python
##pie-Chart, showing top 3 countries
plt.pie(country_values[:3],labels=country_name[:3],autopct="%1.2f%%")
```

```python
## we can observe that India has the 94% of recorded in Zomato followed by USA and UK.Which make it the India centric market.
```

```python
ratings=final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:"Rating count"})
```

```python
##Observatios:
#1. when rating is between 4.5 to 4.9 ----> excellent
#2. when rating is between 4.0 to 4.4 ----> very good
#3. when rating is between 3.5 to 3.9 ----> good
#4. when rating is between 2.5 to 3.4 ----> average
#5. when rating is between 1.8 to 2.4 ----> poor
#6. when rating is 0 --------------------> no ratings
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize'] = (12,6)
sns.barplot(x="Aggregate rating",y="Rating count",data=ratings)
```

```python
sns.barplot(x="Aggregate rating",y="Rating count",hue="Rating color",data=ratings,palette=['white','red','orange','yellow','green','darkgreen'])
```

Observation's:
1. Not rated count Is Maximum of all.
2. Maximum number of rating's are in the range of 3.0 -3.8 which shows the performance of the Hotel is n avegrage range.
3. The average rating in Zomato is 3.4 i.e Good.

```python
##count plot
sns.countplot(x="Rating color",data=ratings,palette=['blue','red','orange','yellow','green','green'])
```

This is basically shows the frequency of rating that comes under the same color code
As white have 0.0 and only 1 row
And the same way orange have 10 row in the data set thats say the its a wide range that shows that average score of the Hotel's.
